# Beam deconvolved data

```python
def return_deconvolved_data(fname,bl,nside):
    tqu=h.read_map(fname,(0,1,2),dtype=np.float64,verbose=False)*1e6
    alm=h.map2alm(tqu,lmax=lmax+lpad)
    invbl=np.ones_like(bl) ; invbl[2:]=1./bl[2:]
    alm[0]=h.almxfl(alm[0],fl=invbl)
    alm[1]=h.almxfl(alm[1],fl=invbl)
    alm[2]=h.almxfl(alm[2],fl=invbl)
    dec_tqu=h.alm2map(alm,pol=True,nside=nside)
    return dec_tqu
    
data={}
data["SMICA"]=return_deconvolved_data(fname_smica,beam["SMICA"],nside)
data["143 GHz"]=return_deconvolved_data(fname_143,beam["143 Ghz"],nside)
```

# Gaussian Likelihood

```python
def calc_u95(Clobs,Nl,fsky,Lmax,Arng=[-5.5,-3],sampling=500):
    from scipy.integrate import romberg
    from scipy.interpolate import interp1d
    Avals=np.logspace(Arng[0],Arng[1],sampling)
    L=arange(Lmax+1)
    clthry=np.ones_like(L) ; clthry[1:]=2.*np.pi/(L[1:]*(L[1:]+1.))
    Avals=np.logspace(Arng[0],Arng[1],sampling)
    Avals=np.append(0,Avals)
    chi2=np.zeros_like(Avals)
    for i,A in enumerate(Avals):
        chi2[i]=bire_lkl(A,Clobs,Nl,clthry,fsky,Lmax)
    
    chi2=chi2/min(chi2)
    fnpdf=interp1d(Avals,np.exp(-chi2),kind="cubic",fill_value="extrapolate")
    norm=romberg(fnpdf,min(Avals),max(Avals))
    fnpdf=interp1d(Avals,np.exp(-chi2)/norm,kind="cubic",fill_value="extrapolate")
    cpdf=np.zeros_like(Avals[2:])
    for i,A in enumerate(Avals[2:]):
        cpdf[i]=quad(fnpdf,min(Avals),A)[0]

    fncpdf=interp1d(cpdf,Avals[2:],kind="cubic",fill_value="extrapolate")
    return fncpdf(0.95)

def bire_lkl(A,Clobs,Nl,clthry,fsky,Lmax):
    L=np.arange(Lmax+1)
    var=(2./((2.*L+1)*fsky))*Nl[:Lmax+1]**2.
    chi2=np.sum((Clobs[1:Lmax]-Nl[1:Lmax]-A*clthry[1:Lmax])**2/var[1:Lmax])
    return chi2
```

```python
stat=collections.OrderedDict()
stat["mask"]=[]
# stat["fsky"]=[]
stat["lmax"]=[]
for k in data.keys():
    stat[k]=[]

for msk in masks.keys():
    for lmax in lmax_list:
        stat["mask"]=stat["mask"]+[msk]
        stat["lmax"]=stat["lmax"]+[lmax]
        for k in data.keys():
            CL=result[k][msk][lmax]["CL"]
            NL=result[k][msk][lmax]["NL"]
            fsky=result[k][msk][lmax]["fsky"]
            try:
                u95=calc_u95(CL,NL,fsky,Lmax)
                stat[k]=stat[k] + [u95]
            except:
                print("Failed : ",msk,lmax,k)
                stat[k]=stat[k] + [-1.]
```

```python
birestat=pd.DataFrame(stat)
birestat.set_index(["mask","lmax"],inplace=True)
for col in birestat.columns:
    birestat[col]=birestat[col].apply(lambda x: '%.2E' % x)
# birestat.index.format(name="fsky",formatter=lambda x: '%.2f' % x)
birestat=birestat.style.set_caption(r"95% upper bounds on $A_{\rm CB}$ from Planck")
birestat=birestat.background_gradient()
filename=figpath + "bire_stat.jpeg"
dfi.export(birestat,filename,chrome_path=None)
```
