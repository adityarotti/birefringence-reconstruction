```python
import os
import healpy as h
import collections
from calc_EB_rec_noise.fast_calc_eb_rec_noise import calc_eb_rec_noise
from calc_TB_rec_noise.fast_calc_tb_rec_noise import calc_tb_rec_noise
import eb_rec
import tb_rec
%pylab notebook
```

```python
clthry=np.loadtxt("../../T2P_leakage/planck_2015_cmb/planck_2015_cmb_totCls.dat")
clthry=np.loadtxt("../../T2P_leakage/planck_2015_cmb/planck_2015_cmb_lensedCls.dat")

maxell=3500
ell=clthry[:maxell+1,0] ; fl=ell*(ell+1)/(2.*pi)
cltt=clthry[:maxell+1,1]/fl ; cltt=append([0.,0.],cltt)
clee=clthry[:maxell+1,2]/fl ; clee=append([0.,0.],clee)
clbb=clthry[:maxell+1,3]/fl ; clbb=append([0.,0.],clbb)
clte=clthry[:maxell+1,4]/fl ; clte=append([0.,0.],clte)
ell=append([0.,1,],ell) ; fl=ell*(ell+1)/(2.*pi)
clthr=[cltt,clee,clbb,clte]

nside=512
lmax=3*nside
```

```python
Lmax=1024
L=np.arange(Lmax+1)
NLeb=calc_eb_rec_noise(clee,clee,clbb,2,1024,Lmax)
# NLtb=calc_eb_rec_noise(clte,clee,clbb,2,1024,Lmax)
NLtb=calc_tb_rec_noise(clte,cltt,clbb,2,1024,Lmax)

figure()
plot(L,NLeb,"r-")
plot(L,NLtb,"--")
loglog()
```

```python
cmb=h.synfast(clthr,nside,new=True,pol=True,verbose=False,lmax=3*nside)
clcmb=h.alm2cl(h.map2alm(cmb,pol=True,lmax=lmax))
```

```python
eb=eb_rec.opt_eb_qe(cmb,clthr,1024,Lmax)
eb.eb_reconstruct()

# cl=h.alm2cl(h.map2alm(eb.rec_alpha,lmax=Lmax))
figure()
plot(eb.Cl_rec_alpha)
plot(NLeb)
# plot(cl,"k--")
loglog()

h.mollview(eb.rec_alpha)
```

```python
tb=tb_rec.opt_tb_qe(cmb,clthr,1024,Lmax)
tb.tb_reconstruct()

cl=h.alm2cl(h.map2alm(eb.rec_alpha,lmax=Lmax))
figure()
plot(tb.Cl_rec_alpha)
plot(tb.N_L)
plot(NLtb,"--")
# plot(cl,"k--")
loglog()

h.mollview(tb.rec_alpha)
```

```python
ell=np.arange(Lmax+1).astype(np.float64) ; cl_alpha=np.zeros_like(ell)
cl_alpha[1:]=1e-4*ell[1:]**-1.
alpha=h.synfast(cls=cl_alpha,nside=nside,verbose=False)
```

```python
obs=np.zeros_like(cmb)
obs[1]=cmb[1] - alpha*cmb[2]
obs[2]=cmb[2] + alpha*cmb[1]
```
